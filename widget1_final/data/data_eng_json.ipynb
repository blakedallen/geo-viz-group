```python
df = pd.read_csv('test3_pct.csv')
df = df.drop(['Year'], axis =1)
df['No'] = np.arange(len(df))
df.head()
```

```python
def data_transform(df, row):
    """data transform to key value pair
    """
    ls = ['Agriculture', 'Bunker Fuels', 'Industry',
       'Land-Use Change and Forestry', 'Waste', 'Buildings',
       'Electricity & Heat', 'Fugitive from energy production',
       'Manufacturing/Construction energy', 'Other Fuel Combustion',
       'Transport']
    
    value = list(df.iloc[row])
    res = []
    for i in range(len(ls)):
#         out = {'label': ls[i], 'value': str(value[i])}
        out = dict({'label': ls[i], 'value': str(value[i])})
        res.append(out)
    return(res)
```

```python
data_final = df.copy()
data_final.head()
```

```python
# converting
data_final['amount'] = data_final['total'].apply(lambda x: str(x))
data_final['data'] = data_final['No']\
    .apply(lambda x: data_transform(df, x))

data_final = data_final.drop([
    'Agriculture', 'Bunker Fuels', 'Industry',
       'Land-Use Change and Forestry', 'Waste', 'Buildings',
       'Electricity & Heat', 'Fugitive from energy production',
       'Manufacturing/Construction energy', 'Other Fuel Combustion',
       'Transport', 'No', 'total'
], axis = 1)
```
